```python
import random
from scipy import stats 
import matplotlib.pyplot as plt
```

```python
lista = sorted([random.randint(1,100) for i in range(10000)])
```

```python
print(lista)
```

```python
#analisar normalidade pelo grfico
plt.hist(lista)
plt.title("Dados")
```

```python
#analisar normalidade pelo grfico
plt.bar(5,6)
normal=[5,5,5,5,5,5]
for i in range(5):
    plt.bar(i,i+1)
    plt.bar(10-i,i+1)
    normal+=[i for j in range(i+1)]
    normal+=[10-i for j in range(i+1)]
    
normal.sort()    
print(normal)
plt.title("Normal")
plt.show()
plt.hist(normal)
plt.show()
```

```python
fig, ax = plt.subplots()
stats.probplot(normal, fit= True, plot=ax)
plt.show()
```

```python
#distribuio no normal
fig, ax = plt.subplots()
stats.probplot(lista, fit= True, plot=ax)
plt.show()
```

```python
#Est normalmente distribuido
stats.shapiro(normal)
```

```python
#No est normalmente distribuido
stats.shapiro(lista)
```
