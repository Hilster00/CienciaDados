```python
import pandas, numpy, random
import matplotlib.pyplot as plt
from yellowbrick.regressor import ResidualsPlot
from sklearn.linear_model import LinearRegression
```

```python
dados1=numpy.array(list(range(100)))
dados2 = numpy.array([i+10 - random.randint(0,5) for i in range(100)])
dados = pandas.DataFrame({"a":dados1,"b":dados2})
```

```python
temp1 = dados.loc[:, "a"].values
temp2= dados.loc[:,"b"].values
temp1 = temp1.reshape(-1, 1)
```

```python
modelo = LinearRegression()
modelo.fit(temp1,dados["b"])
```

```python
modelo.intercept_
```

```python
modelo.coef_
```

```python
plt.plot(temp1, modelo.predict(temp1), color="red")
plt.scatter(dados["a"],dados["b"])
plt.show()
```

```python
previsao1 = lambda i: modelo.intercept_ + modelo.coef_ * i
previsao2 = lambda i: modelo.predict([[i]])
```

```python
previsao1(15)
previsao2(15)
```
